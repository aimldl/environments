## 1. Introduction
### 1.1. 분류 (Classification)
분류는 Categorization이라고 불리기도 하는데, 입력 데이터를 카테고리 별로 나누는 작업입니다. Classification에서는 카테고리 대신 Class라고 부릅니다.

## 1.2. 이진 분류 (Binary Classification) vs. 다중 클래스 분류 (Multi-Class Classification)
위의 경우처럼 2개의 클래스가 있을 때 이진 분류 (Binary Classification)라고 하고, 이진 분류를 하는 문제를 이진 분류 문제 (Binary Classification Problem)이라고 합니다. 2개 이상의 클래스로 분류하는 것과 해당 문제를 각각 다중 클래스 분류 (Multi-Class Classification), 다중 클래스 분류 문제 (Multi-Class Classification Problem)라고 합니다. 예를 들어 3개의 클래스가 있는 경우, 3개의 클래스를 가지는 다중 클래스 분류 (Multi-Class Classification with Three Classes)라고 합니다.

| Classification             | Number of Classes |
|----------------------------|-------------------|
| Binary Classification      | 2                 |
| Multi-Class Classification | More than 2       |


### 1.3. 텍스트 분류 (Text Classification)
Text는 (모든 형태의) 글을 뜻합니다. 그러므로 텍스트 분류 문제 (Text Classification Problem)는 입력 데이터가 글인 것을 분류하는 문제입니다.

### 문서 분류 (Document Classification)
문서 분류기 (Document Classifier)의 입력은 책에서 텍스트만 발췌한 내용이고, 출력은 이 텍스트가 서핑에 관한 것인지 요리에 관한 것입니다. 도서관의 사서 (Librarian)가 책의 텍스트를 훑어보고 이 책의 카테고리가 서핑인지 요리인지 분간하듯이, 분류기 (Classifier)가 입력된 문서들의 종류 (Class)를 추정 (Estimate)합니다.

| A Document | Classification        |  Class 0  |  Class 1 |
|------------|-----------------------|-----------|----------|
| A Document | Binary                | Surfing   | Cooking  |

문서 분류 문제는 분류하고자 하는 문서의 종류 (Class)에 따라 이진 분류 또는 다중 분류 문제가 될 수 있습니다. 주어진 문서가 서핑 (Surfing)에 관한 것인지 요리 (Cooking)에 관한 것인지, 2개로 나뉘면 이진 분류 문제입니다만 더 많은 종류 (Class)로 변경하면 다중 분류 문제가 됩니다. 예를 들면, 음악이라는 분류기준 혹은 종류를 더하면 3개의 클래스를 가지는 다중 분류 문제 입니다.

| A Document | Classification        | Class0  |  Class 1  |  Class 2 |
|------------|-----------------------|---------|-----------|----------|
| A Document | Multi-Class           | Surfing | Cooking   | Music    |

### 스팸메일 분류/ 스팸 필터 (Spam Filter)
스팸 필터는 입력으로 이메일을 받아서, 출력으로 스팸이냐 아니냐를 분류해주는 분류기입니다. 즉, 메일을 일반 메일과 스팸메일으로 자동 분류하는 문제입니다.

스팸 필터링은 스팸이냐 아니냐 2가지로 나뉘어지므로 이진 분류 문제입니다.

| A Document | Classification | Classes          | Class 0  | Class 1 |
|------------|----------------|------------------|----------|---------|
| An Email   | Binary         | Spam or Not Spam | Not Spam | Spam    |

스팸 메일로 분류된 메일의 예입니다.

첫 번째 메일의 내용

"Report not spam" 혹은 "Report spam"버튼을 누르면 해당 메일이 라벨링 (labeling) 혹은 태깅 (Tagging)됩니다.

### 감정 분류/ 감석 분석 (Sentiment Analysis)
주어진 글이 긍정적인지 부정적인지 판단하는 문제입니다. 글에 대한 감정을 더 세분화해서, 금정, 중립, 부정 혹은 더 많은 범주로도 분류할 수 있습니다.

#### 감성 분석의 의미
감성 분석은 오피니언 마이닝 (Opinion Mining) 또는 감정 AI (Emotion AI)로 불리기도 합니다. 감성 분석는 표현이 덜 직관적이므로 이름만으로 무엇을 하는지 이해하기 힘들 수 있습니다. 굳이 표현하자면 텍스트 형태로 된 글에 나타난 감정을 분류하는 문제입니다. 예를 들면, 영화 A에 대한 댓글을 보고 글쓴이가 이 영화를 "좋아한다" "싫어한다"는 감정을 알 수 있으므로 "좋음" "나쁨"으로 분류할 수 있습니다. 

Opinion Mining은 한국로도 오피니언 마이닝이라고 하는 경우가 일반적입니다. (Mining에 적합한 의역을 찾지 못해서 그런 것으로 추측합니다.) 의견 채굴 (=알아내기) 혹은 오피니언 마이닝은 데이터 마이닝 (Data Mining)을 텍스트 데이터에 대해서 한 결과 전체 입력에 대한 의견을 찾아내는 문제입니다. 예를 들면, 영화 A에 대한 댓글을 보고 글쓴이가 이 영화를 "좋아한다" "싫어한다"는 의견을 알아내서 "좋음" "나쁨"으로 분류할 수 있습니다. 그리고 많은 사람의 의견을 알아내면 이 영화에 대한 여론을 알 수도 있어집니다.

앞선 두 예와 같이 감성 분석과 오피니언 마이닝, 두 분야는 세세한 차이는 있지만 유사한 점이 많습니다. 감성 분석과 오피니언 마이닝의 핵심은 텍스트에 나타난 감정을 분류하는 텍스트 분류 문제입니다. 그러므로 감정 AI로 불리기도 합니다.

감성 분석은 다양한 종류의 텍스트에 적용될 수 있습니다. 예를 들면, 트위터 (Twitter)의 댓글 모음집을 입력으로 받거나, 영화 리뷰 모음집 (IMDB Movie Review)을 입력받을 수 있습니다. 두 경우 모두 텍스트 데이터입니다. 이런 텍스트 입력을 받아서 (글쓴이의) 감정상태를 출력합니다. 감정 상태가
* 2가지로 표현되면 좋다 (Positive)-나쁘다 (Negative),
* 3가지로는 좋다 (Positive)-보통 (Neutral)-나쁘다 (Negative),
* 5가지로는 아주 좋다 (Very Good)-좋다 (Good)-보통 (Neutral)-나쁘다 (Bad)-아주 나쁘다 (Very Bad)

와 같이 표현할 수 있습니다.

### 1.5. 용어 정의
다음 단계로 넘어가기 전에 Text를 다룰 때 일반적으로 쓰이는 용어를 몇 가지 정의하겠습니다. 일단 영어를 기준으로 합니다.

* 단어 (Word): (영어에서) 하나의 단어는 White space로 나눌 수 있습니다. 
* 문장 (Sentence): 여러개의 단어가 모여서 하나의 문장을 구성합니다. 
* 문서 (Document): 여러개의 문장이 모여서 하나의 문서를 구성합니다.

단어와 문장의 정의는 일반적인 경우와 동일하나, 문서의 정의가 다른 점에 유의하세요. 이 3가지 단위의 관계는 

\begin{equation*}
Word \subseteq Sentence \subseteq Document
\end{equation*}

문서는 하나의 문장이나 하나의 단어로 구성될 수도 있습니다.

### 1.6. Formal Definition of Text Classification

\begin{equation*}
\hat{c} \leftarrow \gamma(d) \quad where \quad C={c_1, c_2, ..., c_j}
\end{equation*}

```
Given a representation of a document d,
determine the category of d
from a fixed set of classes C = {c1, c2, ..., cj}
where r(d) is a classification function.
```
