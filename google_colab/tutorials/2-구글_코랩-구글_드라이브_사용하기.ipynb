## 구글 드라이브

구글 드라이브 계정에서 데이터를 코랩 노트북으로 가져올 수 있습니다.
GitHub 등 여러 다른 소스에서도 데이터를 가져올 수 있습니다.

자세한 내용은 [데이터로 작업하기](https://colab.research.google.com/notebooks/intro.ipynb?hl=ko#scrollTo=-Rh3-Vt9Nev9)를 참고할 수 있습니다.


### 3.1. 구글 드라이브 마운팅
구글 드라이브 마운팅에 관한 내용은 [데이터 로드: 드라이브, 스프레드 시트, Google CLoud Storage (영문:External data: Local Files, Drive, Sheets, and Cloud Storage)](https://colab.research.google.com/notebooks/io.ipynb)을 참고했습니다.

#### 3.1.1. 요약
`google.colab` 패키지를 이용해서 구글 드라이브를 마운트하면

```python
from google.colab import drive

drive.mount('/content/drive')
```

아래 경로로, 구글 드라이브 및 코랩 노트북의 디렉토리에 접근할 수 있습니다.

* `Drive > My Drive` -> `/content/drive/MyDrive`
* `Drive > My Drive > Colab Notebooks` -> `/content/drive/MyDrive/'Colab Notebooks'`

#### 3.1.2. 구글 드라이브 마운팅 하기
코드 셀에서 drive를 import하고 `/content/drive`를 마운트하면, 아래와 같이 URL과 `authorization code`를 입력하는 박스창이 나옵니다.

python_google_colab_package-1.png

URL를 클릭해서 authorization code를 발급받습니다. 구글 드라이브에 접속하는 권한을 허용하고 하면, 아래와 같이 authorization code가 생성됩니다. 오른쪽에 있는 복사 버튼을 클릭해서 authorization code를 복사합니다.

박스창에 복사한 authorization code를 붙여넣고 엔터키를 입력합니다.

`/content/drive`에 마운트 됐다는 메세지가 나오면 구글 드라이브가 성공적으로 마운트 되었습니다.

python_google_colab_package-6.png

#### 3.1.3. 마운트된 후에 구글 드라이브의 위치
디렉토리의 위치는 `/content/drive/MyDrive`입니다. 웹브라우저에서 보이는 계층도 `Drive > My Drive`임을 알 수 있습니다.

google_drive-my_drive-colab_notebooks.png

그러면 `Drive > My Drive > Colab Notebooks`의 디렉토리는 `/content/drive/MyDrive/'Colab Notebooks'`입니다.
