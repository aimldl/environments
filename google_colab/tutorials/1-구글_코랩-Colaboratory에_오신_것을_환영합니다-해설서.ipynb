## 코랩 노트북 (Colab Notebook)
이 노트북은 코랩 노트북으로 작성되었습니다. [코랩 노트북 (또는 코랩 메모장)](https://colab.research.google.com/notebooks/)은 [쥬피터 노트북(Jupyter Notebook)](https://jupyter.org/)과 유사한 환경을 제공합니다. 노트북은 
* 텍스트와 코드를 하나의 문서에서 작성할 수 있는 대화형 환경으로, 크게
  * 글을 작성할 수 있는 텍스트 셀(text cell)과 프로그램 작성할 수 있는 코드 셀(code cell)로 구성됩니다.
* 텍스트 셀은 마크다운(Markdown) 편집기 입니다. 
  * 마크다운은 텍스트를 입력할 때 서식을 간단하게 입력할 수 있도록 해줍니다.
  * 마크다운 문법을 활용하면
    * 제목, 목록, 링크
    * 이미지, 원시HTML, LaTeX 등
    * 을 손쉽게 입력/편집할 수 있습니다.
  * 마크다운 문법은 [MarkDown 사용법 총정리](https://heropy.blog/2017/09/30/markdown/)를 참고하세요.
* 코드 셀은 리눅스 가상환경(virtual environment)을 제공합니다.
  * 많이 쓰이는 연산환경이 이미 설정된 코드/명령어 해석기로
    * 파이썬 해석기(Python Interpreter)
    * (공통적으로 많이 쓰이는) 파이썬 외장함수
    * 리눅스 쉘
    * GPU
  * 등을 별도로 설치하지 않고 사용할 수 있으며
    * 파이썬 외장함수
    * 리눅스 명령어
  * 를 추가로 설치할 수 있습니다.

## 텍스트 셀: 마크다운 문법 (markdown syntax) 소개
텍스트 셀에 텍스트를 입력할 때 #,* 등과 띄워쓰기를 써서 간단하게 서식을 정할 수 있습니다. 위에 작성된 글의 일부는 #와 *를 적절히 띄워써서 아래처럼 입력되었습니다.

```text

* 텍스트와 코드를 하나의 문서에서 작성할 수 있는 대화형 환경 ...
  * 글을 작성할 수 있는 텍스트 셀(text cell)과 ...
* 텍스트 셀은 마크다운(Markdown) 편집기 입니다. 
  * 마크다운은 텍스트를 입력할 때 서식을 간단하게 입력할 수 있도록 해줍니다.
  * 마크다운 문법을 활용하면
    * 제목, 목록, 링크
    * 이미지, 원시HTML, LaTeX 등
    * 을 손쉽게 입력/편집할 수 있습니다.
```

### 데이터 과학(Data Science)


* 파이썬 외장함수를 쓰려면 `pip install` 명령어 등을 써서 설치하는 등의 환경설정이 필요합니다.
* 구글 코랩에는 일반적으로 자주 쓰는 외장함수가 포함되므로, 별도의 환경설정이 불필요합니다.

아래는 외장함수인 `Numpy`를 사용해서 데이터를 임의로 생성하고, `Matplotlib`를 써서 생성된 데이터를 시각화하는 예입니다.

````
import numpy as np
from matplotlib import pyplot as plt

ys = 200 + np.random.rand(100)
x = [x for x in range( len(ys)) ]

plt.plot( x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys >195), facecolor='g', alpha=0.6)

plt.title( 'Sample Visualization' )
plt.show()
````
