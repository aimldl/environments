# 3. 구글 코랩에서 지원되지 않는 파이썬 기본 문법

* 구글 코랩 코드 셀 (Code cell)의 파이썬 인터프리터에서 지원하지 않아서 이상 동작하는 파이썬 기본 문법이 몇 가지 있습니다.
* 동일한 명령어를 리눅스 터미널의 파이썬 인터프리터에 입력하면 정상 동작합니다.
  * 우분투 리눅스 20.04에 아나콘다 가상환경이 깔린 파이썬 3.8.5 인터프리터를 사용했습니다.

## 여러 줄의 문자열을 쓰고 싶을 때, `"""` 뒤에 `\`를 붙일 경우
[파이썬 자습서](https://docs.python.org/ko/3/tutorial/index.html) > [3.1.2. 문자열](https://docs.python.org/ko/3/tutorial/introduction.html#strings)의 관련 부분을 발췌한 내용입니다.

> 문자열 리터럴은 여러 줄로 확장될 수 있습니다.
>
> 한 가지 방법은 삼중 따옴표를 사용하는 것입니다:
>
> """...""" 또는 '''...'''. 줄 넘김 문자는 자동으로 문자열에 포함됩니다.
>
> 하지만 줄 끝에 \ 를 붙여 이를 방지할 수도 있습니다.

* `"""` 뒤에 `\`를 붙이면 줄넘김 문자를 방지할 수 있다고 설명되어 있습니다.

### 이상 동작
* [파이썬 자습서](https://docs.python.org/ko/3/tutorial/index.html)의 내용대로 입력하면 이상 동작합니다.
  * 예상과 다르게 출력의 첫 줄이 넘어갑니다.

오히려 `\`를 붙이지 않을 때, 줄이 넘어가지 않습니다.

### 정상 동작
* 우분투 리눅스 터미널의 파이썬 인터프리터에선 정상 작동합니다.
* `"""` 뒤에 `\`를 붙였을 때, 줄 넘김 없이 출력됨을 알 수 있습니다.

## 함수 정의 부분에서 위치 전용 매개 변수를 지정하기 위해 `/`를 붙일 경우
[파이썬 자습서](https://docs.python.org/ko/3/tutorial/index.html) > [4.7.3. 특수 매개 변수](https://docs.python.org/ko/3.9/tutorial/controlflow.html#special-parameters) > [4.7.3.2. 위치 전용 매개 변수](https://docs.python.org/ko/3.9/tutorial/controlflow.html#positional-only-parameters)의 관련 부분을 발췌한 내용입니다.

> 특정 매개 변수를 위치 전용으로 표시할 수 있습니다.
>
> 위치 전용이면, 매개 변수의 순서가 중요하며, 키워드로 매개 변수를 전달할 수 없습니다.
>
> 위치 전용 매개 변수는 / (슬래시) 앞에 놓입니다.

[4.7.3.4. 함수 예제](https://docs.python.org/ko/3.9/tutorial/controlflow.html#function-examples)의 2개의 예제에서 에러가 발생합니다.

### 예제 1: `def pos_only_arg(arg, /):`

#### 이상 증상
* 매개 변수를 구분하기 위해, `pos_only` 다음에 `/`를 넣었을 때 `SyntaxError`가 발생합니다.
* 공식 [파이썬 자습서](https://docs.python.org/ko/3/tutorial/index.html) 에 문법 오류가 포함되어 있는 것일까요?

> 두 번째 함수 pos_only_arg는 함수 정의에 /가 있으므로 위치 매개 변수만 사용하도록 제한됩니다:

라는 내용 다음에 실행되는 명령어
```python
pos_only_arg(1)
```
또한 당연히 에러가 발생합니다. 함수를 정의할 때 에러가 발생했으므로, 함수 정의가 안 되어 있어서 발생하는 에러입니다.

#### 정상 동작
* 우분투 리눅스 터미널의 파이썬 인터프리터에선 정상 작동합니다.
* 함수의 매개 변수 정의하는 부분에서 `/`를 써서, 위치 전용임을 명시할 수 있습니다.
